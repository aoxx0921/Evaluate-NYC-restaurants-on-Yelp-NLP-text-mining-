```python
######### 660 project ##########

import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

page = requests.get("https://www.opentable.com/promo.aspx?currentview=list&promoid=69&ref=412&size=100\
&sort=Popularity&covers=2&dateTime=2018-02-20+19%3A00&metroId=8") # send a get request to the web page rows=[]

#status_code 200 indicates success
#a status code>200 indicates a failure 

if page.status_code==200:
    soup = BeautifulSoup(page.content, "html.parser")
    
    # find a block with id='seven-day-forecast-body'
    # follow the path down to the div for each period
    
    divs = soup.select("div#search_results div.content-section-list div.result")
    print(divs)
```

```python
for idx, div in enumerate(divs):
    print(idx, div)
```

```python
# initiate the variable for each period
title = None
recommend = None
cuisine = None

rows = []

for idx, div in enumerate(divs):
    

    # get title
    p_title = div.select("span.rest-row-name-text")

    if p_title!=[]:
        title=p_title[0].get_text()

    # get description
    p_recommend = div.select("span.recommended-text")
    if p_recommend!=[]:
        recommend = p_recommend[0].get_text()

    # get temperature
    p_cuisine = div.select("span.rest-row-meta--cuisine")
    if p_cuisine!=[]:
        cuisine = p_cuisine[0].get_text()

    # add title, description, and temperature as a tuple into the list

    rows.append((title, recommend, cuisine))
    
    print((title,recommend,cuisine))
```
